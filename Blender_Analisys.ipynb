## Filling the NANs using ITEM_ID
If we simply cleared all the NaN we would lose 85% of the information so we need to see how to solve the NAN issue for each column. The information we have is the ITEM_ID, we will try to get the information for each type of item and fill it in the missing value. I'm assuming that the ITEM_ID is the product type so for the same product type the characteristics would be the same. This logic would not be used for price because this can change over time

```python
item_IDS =  df.ITEM_ID.unique()
print(len(item_IDS))
total_updated = 0
for id in item_IDS:
    filtered = df.loc[df['ITEM_ID'] == id].head().dropna()
    if(filtered.size!=0):
       total_updated +=1
    
    
print(total_updated)
```

Observing the Data above we can see that only 31 products have all the information so we need to find another strategy to fill in the NaN. Also we can see if we would drop the NAN rows we would have information only about 31 products

The First step will be for each column that has a NaN value we will create another categoric column, that will indicate for us if the product has or not the information related with the column. This will help the algoritm to understand if the information is relevant on the classification. After that the NAN will be replaced by zeros. 

```python
columns_with_missing_info = []
columns_with_all_info = []
for column in df.columns:
    if(df[column].isna().sum()>0):
        columns_with_missing_info.append(column)
        new_column = column+"_CONTAINS_INFO"
        df[new_column] = df[column].isna()
    else:
        if (column.find("_CONTAINS_INFO")==-1):
            columns_with_all_info.append(column)

print(columns_with_missing_info)
print(columns_with_all_info)
```

We have 6 numerical features missing information 
- ALTURA, CAPACIDADE_(L),LARGURA,PESO,PROFUNDIDADE,TEMPO_GARANTIA,ITEM_PRICE

We have 5 categorical features missing information 
- COMPOSICAO, COR, FORMATO, MARCA, PARA_LAVA_LOUCAS, PARA_MICRO_ONDAS

We are ignoring SESSION_ID for now because that information will be used later. For the missing information on the numerical columns we will use a LINEAR REGRESSION to infer de value for NAN. So in the following steps we will create the linear regression for each missing of those. But to apply that we will use the DataFrame with all the NaN removed and after that we will enconde the categorical columns using SKLearn build in feature.

```python
categorical_cols= ['COMPOSICAO', 'COR', 'FORMATO','MARCA', 'PARA_LAVA_LOUCAS', 'PARA_MICRO_ONDAS','TEM_FERRO_FUNDIDO', 'TEM_GRELHA','TEM_TAMPA', 'TIPO_PRODUTO', 'TIPO_WOK']
cleared_df = pd.get_dummies(cleared_df, columns=categorical_cols)
print(cleared_df.head())
final_cols =  [item for item in cleared_df.columns if item not in categorical_cols]
```

```python
df_for_reg = cleared_df.filter(final_cols)
df_for_reg = df_for_reg.drop(columns=['SESSION_ID'])
df_for_reg.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split  
import numpy as np
regressors = {}
```

```python
for column in ['ALTURA', 'CAPACIDADE_(L)','LARGURA','PESO','PROFUNDIDADE','TEMPO_GARANTIA','ITEM_PRICE']:
    regressor = LinearRegression()  
    y = df_for_reg[column].values
    x = df_for_reg.loc[:, df_for_reg.columns != column].values
    X_train, X_test, y_train, y_test = train_test_split(x, y)  
    regressor.fit(X_train, y_train)  
    result =regressor.score(X_test,y_test)
    print(column+" Accuracy: " + str(result))
    regressors[column] = regressor
```

We can se that error is very low for all the regressors, so we can use than to clean the NaNs of the numerical columns.
Now we train a simple NN to each categorical feature with missing information. And use then the same way of the regresssors.

```python
from sklearn.neural_network import MLPClassifier
for column in cleared_df.columns:
    cleared_df[column] = cleared_df[column].astype(float)
classifiers = {}
for column in categorical_cols:
    classifier = MLPClassifier(activation='relu', hidden_layer_sizes=(5, 2), learning_rate='constant',
       learning_rate_init=0.001, max_iter=1000, momentum=0.9,
       nesterovs_momentum=True, verbose=False)  
    filter_col = [col for col in cleared_df if col.startswith(column)]
    y = cleared_df[filter_col].values
    x = cleared_df.drop(columns=filter_col).values
    X_train, X_test, y_train, y_test = train_test_split(x, y)  
    classifier.fit(X_train, y_train)  
    result =classifier.score(X_test,y_test)
    print(column+" Accuracy: " + str(result))
    classifiers[column] = classifier
```

We can se that error is very low for all the regressors, so we can use than to clean the NaNs of the categorical columns.
